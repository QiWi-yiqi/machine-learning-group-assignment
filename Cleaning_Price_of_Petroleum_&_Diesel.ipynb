check outlier

```python
import numpy as np

outlier_counts = {}

for column in df.select_dtypes(include=np.number).columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    outlier_counts[column] = len(outliers)

print("Number of outliers per numerical column (using IQR method):")
for col, count in outlier_counts.items():
    print(f"  {col}: {count}")
```

interpolate into daily data

```python
import pandas as pd
import numpy as np

# Read the CSV
df = pd.read_csv("weekly_petroleumdata_cleaned.csv", keep_default_na=False, na_values=[])

# Extract the start date of each week
# The original file 'weekly_carregistered_cleaned.csv' was expected to have a 'week' column.
# Since we are now using 'weekly_petroleumdata_cleaned.csv', which has a 'date' column,
# we should use the 'date' column directly and set it as the index.

df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Explicitly convert relevant columns to numeric type to ensure interpolation works
# Use pd.to_numeric with errors='coerce' to handle potential empty strings or other non-numeric values
for col in ['ron95', 'ron97', 'diesel']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# The original code was intended for a dataset with a 'count' column that needed aggregation.
# The 'weekly_petroleumdata_cleaned.csv' dataset already contains weekly aggregated data for ron95, ron97, and diesel.
# Therefore, resampling to daily and interpolating might not be the desired action here.
# If the goal is to have daily data, then the approach below would be appropriate, but it might not be relevant for this petroleum data.
# For now, I will assume the goal is to prepare the petroleum data for further analysis, not resample it in this specific way.
# If you still want to resample and interpolate this petroleum data, please let me know.

# Keeping the original resampling logic, but acknowledging its potential irrelevance for this specific dataset.
# If you want to analyze the petroleum data directly as weekly data, these lines should be removed or modified.
# For demonstration, I will adapt the resampling to the existing 'ron95' column as an example.
# If you want to interpolate all columns, you can apply this to all relevant columns.

# Example: Resample 'ron95' to daily frequency and interpolate
daily_data = df[['ron95', 'ron97', 'diesel']].resample('D').interpolate(method='linear')

# Reset index to get 'Date' as a column again
daily_data.reset_index(inplace=True)
daily_data.rename(columns={'date': 'Date'}, inplace=True)

print(daily_data)
```

download the dataset

```python
from google.colab import files

# Save the aggregated DataFrame to a CSV file
output_filename_daily = 'interpolated_petroleum_daily_data.csv'
daily_data.to_csv(output_filename_daily, index=False)

# Download the file
files.download(output_filename_daily)
print(f"Interpolated daily DataFrame saved and downloaded as {output_filename_daily}")
```
