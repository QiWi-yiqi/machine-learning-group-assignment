```python
import pandas as pd

# Define the common date range for filtering
start_date = '2017-04-01'
end_date = '2025-10-01'

# Re-load df2 (petroleum data) to ensure correctness, as the kernel state was inconsistent
df2_petroleum = pd.read_csv("Daily Petroleum.csv", keep_default_na=False, na_values=[])
df2_petroleum['Date'] = pd.to_datetime(df2_petroleum['Date'])
df2_petroleum = df2_petroleum[(df2_petroleum['Date'] >= start_date) & (df2_petroleum['Date'] <= end_date)]

# Ensure all original dataframes have their 'Date' column as datetime objects
# (These operations were done in previous cells, but are included for robustness)
df['Date'] = pd.to_datetime(df['Date'])
df1['Date'] = pd.to_datetime(df1['Date'])
df3['Date'] = pd.to_datetime(df3['Date'])

# Perform the merges sequentially
# Merge df (car registered) and df1 (CPI)
merged_data = pd.merge(df, df1, on='Date', how='inner')

# Merge the result with the correctly loaded df2_petroleum
merged_data = pd.merge(merged_data, df2_petroleum, on='Date', how='inner')

# Merge the result with df3 (unemployment rate)
merged_data = pd.merge(merged_data, df3, on='Date', how='inner')

# Display the head and info of the merged DataFrame
print("Merged Data Head:")
print(merged_data.head())
print("\nMerged Data Info:")
print(merged_data.info())
```
