#2017

```python
import pandas as pd

from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("cars_2017.csv", keep_default_na=False, na_values=[])
df
```

```python
print(df['fuel'].unique())
```

```python
allowed_fuel_types = ['petrol']
filtered_df = df[df['fuel'].isin(allowed_fuel_types)]
filtered_df
```

```python
filtered_df = filtered_df.drop(columns=['maker', 'model', 'colour'])
```

```python
display(filtered_df)
```

```python
filtered_df['date_reg'] = pd.to_datetime(filtered_df['date_reg'])
filtered_df['week'] = filtered_df['date_reg'].dt.to_period('W')
monthly_filtered_data = filtered_df.groupby('week').size().reset_index(name='count')
print(monthly_filtered_data)
```

```python
from google.colab import files

csv_file_path = 'weekly_filtered_data_2018.csv'
monthly_filtered_data.to_csv(csv_file_path, index=False)
files.download(csv_file_path)
```

#Interpolate the weekly data into daily data

choosen file

```python
import pandas as pd

from google.colab import files
uploaded = files.upload()
```

show dataset

Interpolate wekly data into daily

```python
import pandas as pd

# Read the CSV
df = pd.read_csv("weekly_carregistered_cleaned.csv", keep_default_na=False, na_values=[])

# Extract the start date of each week
df['start_date'] = df['week'].str.split('/').str[0]
df['start_date'] = pd.to_datetime(df['start_date'])

# Set the start_date as the index
df.set_index('start_date', inplace=True)

# Aggregate duplicate start_date entries by summing their 'count' values
weekly_counts = df.groupby(df.index)['count'].sum().to_frame()

# Resample to daily frequency and interpolate
daily_counts = weekly_counts.resample('D').interpolate(method='linear')

# Reset index to get 'Date' as a column again
daily_counts.reset_index(inplace=True)
daily_counts.rename(columns={'start_date': 'Date'}, inplace=True)

print(daily_counts)
```

download the dataset

```python
from google.colab import files

csv_file_path = 'combined_daily_data_car_registered.csv'
daily_counts.to_csv(csv_file_path, index=False)
files.download(csv_file_path)
```
