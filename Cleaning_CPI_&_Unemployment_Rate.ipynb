interpolate into daily data

```python
import pandas as pd
import numpy as np

# Read the CSV
df = pd.read_csv("CPIAUCSL.csv", keep_default_na=False, na_values=[])

# Convert the 'observation_date' column to datetime and set it as the index
df['observation_date'] = pd.to_datetime(df['observation_date'])
df.set_index('observation_date', inplace=True)

# Explicitly convert the 'CPIAUCSL' column to numeric type, coercing errors
df['CPIAUCSL'] = pd.to_numeric(df['CPIAUCSL'], errors='coerce')

# Resample 'CPIAUCSL' to daily frequency and interpolate
daily_data = df[['CPIAUCSL']].resample('D').interpolate(method='linear')

# Reset index to get 'Date' as a column again
daily_data.reset_index(inplace=True)
daily_data.rename(columns={'observation_date': 'Date'}, inplace=True)

print(daily_data)
```

download as csv file

```python
from google.colab import files

# Save the interpolated daily CPI DataFrame to a CSV file
output_filename_daily_cpi = 'interpolated_CPI_daily_data.csv'
daily_data.to_csv(output_filename_daily_cpi, index=False)

# Download the file
files.download(output_filename_daily_cpi)
print(f"Interpolated daily CPI DataFrame saved and downloaded as {output_filename_daily_cpi}")
```

```python
import pandas as pd
import numpy as np

# Read the CSV
df = pd.read_csv("UNRATE.csv", keep_default_na=False, na_values=[])

# Convert the 'observation_date' column to datetime and set it as the index
df['observation_date'] = pd.to_datetime(df['observation_date'])
df.set_index('observation_date', inplace=True)

# Explicitly convert the 'UNRATE' column to numeric type, coercing errors
df['UNRATE'] = pd.to_numeric(df['UNRATE'], errors='coerce')

# Resample 'UNRATE' to daily frequency and interpolate
daily_data = df[['UNRATE']].resample('D').interpolate(method='linear')

# Reset index to get 'Date' as a column again
daily_data.reset_index(inplace=True)
daily_data.rename(columns={'observation_date': 'Date'}, inplace=True)

print(daily_data)
```
